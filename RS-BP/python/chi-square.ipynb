# The $\chi^2$ distribution

We consider fitting some function $f(x;a_1,a_2, \ldots, a_n)$ to a set of data-points given by the values $(x_i,y_i)$ with an error bar on the $y$-value given by $\sigma_i$. The fit parameters are labeled by the $a$'s, i.e. $a_1,a_2, \ldots, a_n$. The best fit is found by minimizing the least-squares defined as

$$\chi^2 = \sum_i \frac{(y_i - f(x_i;a_1,a_2, \ldots, a_n)^2}{\sigma_i^2}$$

Besides minimizing the value of this $\chi^2$ which gives the best fit of the function to the data we should also assess if the fit is reasonable on statistical grounds. We would expect that 65% of datapoints are only one standard deviation away from the curve and 95% of the data-points are within 2 standard deviations from the curve. This can be quantified if we consider the probability distribution of $\chi^2$. This probability distribution depends on the number of degrees of freedom in the fit. This number of degrees of freedom is equal to the number of datapoints minus the number of fit parameters. 

The easiest least-squares fit is obtained if we consider finding the weighted average. In that case we fit the function $f(x;a_1,a_2, \ldots, a_n) = f(a_1) = a_1 = \bar{y}$. If we only have one datapoint and one fit parameter (the average in this case) there is very little left to do.

**In general, if the number of fit parameters is equal to the number of datapoints you should try to solve the set of equations instead of fitting**. If you have experimental data with an error bar you may find a contradiction, making it impossible to solve the set of equations. You can still find the minimum of $\chi^2$, but having no degrees of freedom you cannot easily assess if this fit is any good. In that case plot the data and your fit. The simplest example is fitting a straight line $y=ax+b$ with $a$ and $b$ as fit parameters to 2 datapoints. In this case you should always find a $\chi^2_{min} = 0$ and the probability to find a value of $\chi^2 \ge 0$ equals 1.

The probabilty density function for $\chi^2$ is given by 

$$P(\chi^2,n) = \frac{2^{-(\frac{n}{2})}}{\Gamma \left( \frac{n}{2} \right)} \chi^{n-2} \mathrm{e}^{-\frac{\chi^2}{2}}$$

where $n$ is the number of degrees of freedom and $\Gamma \left( \frac{n}{2} \right)$ is the Gamma function (https://en.wikipedia.org/wiki/Gamma_function).

For 'large' $n$ this probability density function tends to a Gaussian probability density fucntion with an average $n$ and standard deviation $\sqrt{2 n}$. A quick rule-of-thumb that is often applied is that the normalized value of $\chi^2$ (normalized by dividing by the number of degrees of freedom) should be equal to 1. The standard deviation of this normalized distribution is $\sqrt{2/n}$. For large $n$ it is safe to assume that the probability that you have a good fit is less than 5% if the value is more than 2$\sigma$ away from 1.0.

Python has library functions (https://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.stats.chi2.html) to get the probability density function for the $\chi^2$ distribution. Similar library functions exist for the normal distribution, so it is not so difficult to calculate the probability.

## $\chi^2$ distribution compared to the normal distribution

For a large number of degrees of freedom the $\chi^2$ distribution tends to a normal distribution. For a small number of degrees of freedome the differences are significant. For instance the borderline case of 2 degrees of freedom gives a distribution with a mean of 2 and a variance of 2. For a normal distribution there would be significant probability to obtain negative values for $\chi^2$, something we consider to be impossible since both the function and data points are real-valued.

It is easy to see that approximating the $\chi^2$ distribution with a gaussian is a bad idea for a small number of degrees of freedom. In that case it is better to simply evaluate the probability that the value of $\chi^2$ is larger (or smaller) than the found value. You could use the cummulative density function for this. 

It is interesting to compare this to the quick intuitive guess based on the normal (Gaussian) distribution. The code below allows to evaluate how likely it is to find a certain value of $\chi^2$ using the real distribution and the Gaussian distribution. 

#### You should be able to verify that the Gaussian approximation is
1. Not very accurate for a small number of degrees of freedom and a reasonable value of $\chi^2$ (e.g. equal to the number of degrees of freedom)
2. That the approximation fails in the tails of the distribution. The normal distribtuion drops much quicker and this really matters if you want to use more stringent criteria than 2$\sigma$. For instance the gold standard for resolutely refuting a theory is a 5$\sigma$ probability based on a fit to experimental data.

## The cummaltive density function

The normal distribution with mean $\mu$ and standard deviation $\sigma$ is described by the function

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \exp \left(-\frac{(x-\mu)^2}{2\sigma^2} \right)$$

This distribution has very convenient mathematical properties; it has only a first moment ($\mu$) and second moment ($\sigma^2$), all higher order moments are zero. The Fourier transform of a Gaussian is a Gaussian. Both these properties play a role in the central limit theorem that states that the limiting probability distribution of $N$ independent observables with each on non-Gaussian probability distribution(*) becomes Gaussian in the limit that $N$ goes to infinity.

(*) The underlying distributions should have a well defined first and second moment, i.e. it cannot be the Cauchy distribution (known to physicists as a Lorentzian)

The normal distrbution is a very common distribution so it is useful to develop some intuition for this distribution. For in stance the probability that the value of $x$ is larger than $\mu + 2 \sigma$ should be 2.5%. This probability is defined as

$$ P(x>a) = \int_{a}^\infty \frac{1}{\sigma \sqrt{2\pi}} \exp \left(-\frac{(x-\mu)^2}{2\sigma^2} \right) \mathrm{d}x$$

The answer to this integral has to be obtained by numerical integration or by using tabulated values. Python has a library function for the cummalitive density function, i.e. for the probability

$$ P(x<a) = 1-P(x>a) = \int_{-\infty}^a \frac{1}{\sigma \sqrt{2\pi}} \exp \left(-\frac{(x-\mu)^2}{2\sigma^2} \right) \mathrm{d}x$$

A similar definition holds for the $\chi^2$ value using the probabilty distribution $P(\chi^2,n)$ that depends on the number of degrees of freedom $n$. Note that the value of $\chi^2$ is strictly positive and that the integral thus has different limits, e.g. 

$$ P(\chi^2 < a) = \int_{0}^a P(\chi^2,n) \mathrm{d}\chi^2$$

Python library functions for the cummulative distribution function are given within scipy.stats by norm.cdf for the normal distribution and chi2.cdf for the $\chi^2$ distribution.

```python
x = np.linspace(-5, 5, 100)
xx = np.linspace(-5, -1.5, 30)

plt.figure()
plt.plot(x, norm.pdf(x), ls='solid', color='blue', label='pdf')
plt.fill_between(xx, xx*0.0, norm.pdf(xx), color='blue', alpha=.25)
plt.plot(x, norm.cdf(x), ls='dashed', color='blue', label='cdf')
plt.legend()
plt.xlim(-3,3)
plt.xlabel('x')
plt.ylim(0,1)
plt.ylabel('Probability density')

plt.show()
```

### Example for normal distribution 

Common wisdom says that approximately 65% of the data is within one standard deviation from the average and 95% is within 2 standard deviations from the average. We see from the definition that the normal distribution is symmetric in the value of x. We can thus calculate the probalities $P(x<\mu-\sigma)$ and $P(x<\mu -2 \sigma)$. These should be 37.5% and 2.5% respectively.

Let's find the exact values

```python
print('%6.3f percent of the data is within 1 standard deviation from the mean' % (100.0-200.0*norm.cdf(-1.0)))
print('%6.3f percent of the data is within 2 standard deviations from the mean' % (100.0-200.0*norm.cdf(-2.0)))
```

### Example with the $\chi^2$ distribution

Calculating a similar bound for the $\chi^2$ value is just as easy using the library functions. Let us calculate the probability to find a value of $\chi^2$ that is larger than the mean value for a specific number of degrees of freedom. Let's make a plot of what we try to compute (the blue area) for the case of 3 degrees of freedom.

```python
x = np.linspace(0, 15, 100)
xx = np.linspace(3, 15, 100)

plt.figure()
plt.plot(x, chi2.pdf(x,3), ls='solid', color='blue', label='pdf')
plt.fill_between(xx, xx*0.0, chi2.pdf(xx,3), color='blue', alpha=.25)
plt.xlim(0,15)
plt.xlabel('x')
plt.xticks((0,3,6,9,12))
plt.ylim(0,0.3)
plt.ylabel('Probability density')

plt.show()
```

```python
dof=3 # number of degrees of freedom
chi2value=dof # the value of chi2 from the least squares fit

print('If we assume a normal distribtuion P( chi2 < %d) = %4.3f' % (chi2value, norm.cdf((dof-chi2value)/np.sqrt(2*dof))))

print('\nLeast-squares:\n')

for dof in range(1,20):
    print('For %2d degrees of freedom  P( chi2 < %2d) = %4.3f' %(dof,dof,chi2.cdf(dof,dof)))
```

### Beware of the tail of the distribution

The $\chi^2$ distribution is not Gaussian and this gives some noticable deviations for values 1-2 standard deviations away from the average / expectation value for a low number of degrees of freedom. If we need to take an important decision like discarding a well-accepted piece of theory based on a new observation we should be very sure that it is really improbable. The accepted gold standard is to take the probability 5 standard deviations away from the average as a measure. For a normal distribution the probability than drops below 1 in a million. For a least-squares fit the difference in the tail is quite significant and the fit should be a really poor fit in order to justify this.

```python
fivesigma=norm.cdf(-5,0,1) # level of confidence for the gold standard of five sigma
print(fivesigma)
```

```python
x = np.linspace(0, 50, 100)

plt.figure()
plt.semilogy(x, chi2.pdf(x,10), ls='solid', color='blue')
plt.semilogy(x, norm.pdf(x,10,np.sqrt(20)), ls='dashed', color='red')
plt.semilogy(x, x*0.0+fivesigma, ls='dotted', color='black')
plt.xlabel('$\chi^2$ value')
plt.ylabel('Probability density')
plt.ylim(1E-8,2E-1)

plt.show()
```
