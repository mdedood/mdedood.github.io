# Exercise 2: Determining the gravitational constant

We try to determine the gravitational constant $g$ from an experiment. In the experiment a small metal sphere is released from an adjustable height $h$ and the time to reach the ground is recorded. The sphere is held in place by an electro-magnet and is dropped at time $t = 0$ using a switch that simultaneously starts a measurement of the time. The measurement is stopped when the sphere reaches the ground and thus the time is recorded. There are possible delays in both the moment that the electro-magnet releases the ball and the starting and stopping of the time recording device. 

Repeated measurements for different initial heights result in a set of data that we can compare to the known relation 

$$h = \frac{1}{2} g t^2$$

to determine $g$. We assume that errors in the height are negligible and that a measurement of the time is accurate to 0.01 sec (standard deviation).

The data is stored in the file [timeandheight.dat](timeandheight.dat) and is in comma separated format. We will first read in the data using the numpy loadtxt function.

```python
import numpy as np

t, h = np.loadtxt('timeandheight.dat', delimiter=',', skiprows=5, usecols=(0, 1), unpack=True)

for i in range(len(t)):
    print('h = %4.3f \t t = %4.3f +/- 0.01 s' % (h[i],t[i]))
```

Assume that the sphere really starts to fall at $t = 0$, i.e. that the magnet releases the ball as soon as the button is pressed.

### a) Determine a best estimate (weighted average) of the gravitational constant $g$ and an error bar without using a curve-fit or optimization routine.

It could be useful to use a formula to propagate the error or uncertainty in the time measurement. See https://en.wikipedia.org/wiki/Propagation_of_uncertainty

**Solution:** One possible way to solve the problem would be to rewrite the equation as 
$$g=\frac{2h}{t^2}$$ 
and calculate the value of $g$ and the error bar for each point. To get the error bar we propagate the error 
$${\sigma_g}^2 = \left(\frac{\partial g}{\partial h} \right)^2 {\sigma_h}^2 + \left(\frac{\partial g}{\partial t} \right)^2 {\sigma_t}^2$$
In this exercise the first terms cancels (we assume/use that ${\sigma_h}^2 = 0$). We are thus left with
$${\sigma_g} = \frac{4 h}{t^3} \sigma_t$$
where the error in the time $\sigma_t = 0.01$ s is given.

```python
g = 2*h/t**2
sigma=(4*h/t**3)*0.01 # I compute sigma, not sigma^2 !
for i in range(len(g)):
    print('g = %4.3f +\- %4.3f m/s^2' % (g[i],sigma[i]))

print('\nThe weighted average gives g = %4.3f +\- %4.3f m/s^2' %
      (np.average(g, weights=1.0/sigma**2), np.sqrt(1.0/np.sum(1.0/sigma**2))))
```

### b) Repeat the exercise, but use a curve-fitting tool or least-squares optimization to find a best fit. What is the estimated error on the gravitational constant?

**First solution:** If you call the fit function without using the weight you do get a fit and even an error bar on the fit parameters. If you do not give an errorbar to a fit routine simply use the difference between data and fit squared, i.e. an effective weight of 1. The result is not dimensionless and is useless testing the goodness of fit using the $\chi^2$-value. Note that it is still a 'best fit' to the data; you are still finding the minimum. In this case, the method might be tempting because you can use the original equation and and 'solve' the problem of having error bars in the x-direction by simply ignoring the error bars alltogether. 

In principle one could convert the error bar using an inverse function. This would get you to the same point as the second solution.

```python
# Fit function
def height(t, g):
    "Returns the height h as a fucntion of time it takes to fall"
    return (0.5*g*t**2)

from scipy.optimize import curve_fit
# The actual fit is done by a call to curve_fit. All points have the same error
# so there is no need to specify the weigth per point
popt, pcov = curve_fit(height,t, h)

print('g = %4.3f +/- %4.3f \n' % (popt[0], pcov[0,0]**0.5))

# We will need the minimum value of chi-squared later, so I calculate it here.
# To get the normalized value we need to include the error in this calculation
chisqmin = np.sum((t-height(t,popt[0]))**2)
print('The non-normalized value of chi square = %4.3f for 4 degrees of freedom' % (chisqmin))

# I first plot the 2 sigma confidence interval, then the fit on top and finally the data
tfine=np.linspace(0.,1.2,100) # Use a fine grid to get a smooth curve!
plt.plot(tfine, height(tfine, popt[0]), color='red', ls='solid')
plt.fill_between(tfine, 
                 height(tfine, popt[0]+2*pcov[0,0]**0.5),height(tfine, popt[0]-2*pcov[0,0]**0.5), 
                 color='grey', alpha=0.5)
plt.errorbar(t, h, xerr=0.01, ls='None', marker='o', markersize=5, capsize=7)
plt.xlabel('Time (sec)') # Axes have labels!
plt.ylabel('Height (m)')
plt.ylim((0,6)) 
plt.xlim((0,1.1)) 
plt.xticks((0,0.5,1.0))
plt.show()
```

**Second solution:** Instead of the parabola through the origin we fit a square root function by plotting the time as a function of height. The fit function we look for is 
$$t = \sqrt{\frac{2h}{g}}$$
In this way accounting for the error bar becomes easy and we should be able to calculate a normalized value of $\chi^2$ that we can interpret when we now the degrees of freedom (number of data points minus number of fit parameters).

```python
def time(h, g):
    "Returns the time it takes to fall from a height h"
    return (np.sqrt(2.0*h/g))

popt, pcov = curve_fit(time,h,t)

print('g = %4.3f +/- %4.3f \n' % (popt[0], pcov[0,0]**0.5))

# We will need the minimum value of chi-squared later, so I calculate it here.
# To get the normalized value we need to include the error in this calculation
chisqmin = np.sum((t-time(h,popt[0]))**2/0.01**2)
print('The normalized value of chi square = %4.3f for 4 degrees of freedom' % (chisqmin))

# I first plot the 2 sigma confidence interval, then the fit on top and finally the data
hfine=np.linspace(0.,6.0,100) # Use a fine grid to get a smooth curve!
plt.plot(hfine, time(hfine, popt[0]), color='red', ls='solid')
plt.fill_between(hfine, 
                 time(hfine, popt[0]+2*pcov[0,0]**0.5),time(hfine, popt[0]-2*pcov[0,0]**0.5), 
                 color='grey', alpha=0.5)
plt.errorbar(h,t, yerr=0.01, ls='None', marker='o', markersize=5, capsize=7)
plt.xlabel('Height (m)') # Axes have labels!
plt.ylabel('Time (sec)')
plt.xlim((0,6)) 
plt.ylim((0,1.1)) 
plt.yticks((0,0.5,1.0))
plt.show()
```

It is possible to plot the data and fit as a straight line by taking the square root of the height and plotting the time as a function of square root height. The slope of the line is $2/g$.

The alternative is to plot height as a function of $t^2$. In this case you need to compute the error in $t^2$ instead of in $t$ by multiplying the values by $2t$. The slope of the straight line should be $g/2$.

```python
hfine=np.linspace(0.,6.0,100) # Use a fine grid to get a smooth curve!
plt.plot(np.sqrt(hfine), time(hfine, popt[0]), color='red', ls='solid')
plt.fill_between(np.sqrt(hfine), 
                 time(hfine, popt[0]+2*pcov[0,0]**0.5),time(hfine, popt[0]-2*pcov[0,0]**0.5), 
                 color='grey', alpha=0.5)
plt.errorbar(np.sqrt(h),t, yerr=0.01, ls='None', marker='o', markersize=5, capsize=7)
plt.xlabel('$\sqrt{Height (m)}$') # Axes have labels!
plt.ylabel('Time (sec)')
plt.xlim((0,2.5)) 
plt.ylim((0,1.1)) 
plt.yticks((0,0.5,1.0))
plt.show()
```

The assumption that the ball is released when the button is pressed is not realistic. It usually takes some time for the currents to reach zero. Take this effect into account in your data analysis by introducing an unknown time delay. 

### c) What is the best estimate for the  value of the gravitational constant and what is the error bar in this case?

We adapt the fit function to incorporate the time delay. 
$$h=\frac{1}{2} g (t-t_0)^2$$
In this case I prefer no extra calculation of error bars and use the inverse function $t(h)$ that should be adapted to
$$t=\sqrt{\frac{2}{g}} h + t_0 $$

```python
def time2(h, g, t0):
    "Returns the time it takes to fall from a height h"
    return (np.sqrt(2.0*h/g)+t0)

popt, pcov = curve_fit(time2,h,t)

print('g  = %4.3f +/- %4.3f' % (popt[0], pcov[0,0]**0.5))
print('t0 = %4.3f +/- %4.3f \n' % (popt[1], pcov[1,1]**0.5))

# We will need the minimum value of chi-squared later, so I calculate it here.
# To get the normalized value we need to include the error in this calculation
chisqmin = np.sum((t-time2(h,popt[0],popt[1]))**2/0.01**2)
print('The normalized value of chi square = %4.3f for 3 degrees of freedom' % (chisqmin))

# I first plot the 2 sigma confidence interval, then the fit on top and finally the data
hfine=np.linspace(0.,6.0,100) # Use a fine grid to get a smooth curve!
plt.plot(hfine, time2(hfine, popt[0],popt[1]), color='red', ls='solid')
plt.fill_between(hfine, 
                 time2(hfine, popt[0]+2*pcov[0,0]**0.5, popt[1]),
                 time2(hfine, popt[0]-2*pcov[0,0]**0.5, popt[1]), 
                 color='grey', alpha=0.5)
plt.fill_between(hfine, 
                 time2(hfine, popt[0], popt[1]+2*pcov[1,1]**0.5),
                 time2(hfine, popt[0], popt[1]-2*pcov[1,1]**0.5), 
                 color='grey', alpha=0.5)
plt.errorbar(h,t, yerr=0.01, ls='None', marker='o', markersize=5, capsize=7)
plt.xlabel('Height (m)') # Axes have labels!
plt.ylabel('Time (sec)')
plt.xlim((0,6)) 
plt.ylim((0,1.1)) 
plt.yticks((0,0.5,1.0))
plt.show()

#linear plot
hfine=np.linspace(0.,6.0,100) # Use a fine grid to get a smooth curve!
plt.plot(np.sqrt(hfine), time2(hfine, popt[0],popt[1]), color='red', ls='solid')
plt.fill_between(np.sqrt(hfine), 
                 time2(hfine, popt[0]+2*pcov[0,0]**0.5, popt[1]),
                 time2(hfine, popt[0]-2*pcov[0,0]**0.5, popt[1]), 
                 color='grey', alpha=0.5)
plt.fill_between(np.sqrt(hfine), 
                 time2(hfine, popt[0], popt[1]+2*pcov[1,1]**0.5),
                 time2(hfine, popt[0], popt[1]-2*pcov[1,1]**0.5), 
                 color='grey', alpha=0.5)
plt.errorbar(np.sqrt(h),t, yerr=0.01, ls='None', marker='o', markersize=5, capsize=7)
plt.xlabel('$\sqrt{Height (m)}$') # Axes have labels!
plt.ylabel('Time (sec)')
plt.xlim((0,2.5)) 
plt.ylim((0,1.1)) 
plt.yticks((0,0.5,1.0))
plt.show()
```

### d) Compare your answers for the three cases and discuss the difference.

All 3 methods give a different result. The methods in (a) and (b) ignore a possible offset in the data which explains the difference with the result in (c). The result in (a) is the weighted average of the slope of a straight line through the origin for each data point. The result in (b) is the slope of a straight line through the origin that minimizes a $\chi^2$ value. You get different results depending on whether you fit $t(h)$ or $h(t)$. In (c) you determine the slope of the best fit to get $g$ and you can (always!) reduce the value of $\chi^2$ by introducing an extra fit parameter. ### e) Which of the two models provides a better fit to the data and why?

The normalized values of $\chi^2$ divided by the number of degrees of freedom are 5.9 and 1.3 for the model in (c) and (b), respectively. The standard deviation for the $\chi^2$ distribution should be $\sqrt{2/n}$ when we divide by the number of degrees of freedom that we denote by $n$. 

The models in (b) has 4 degrees of freedom, while the model in (c) has 3 degrees of freedom. This gives a value of $\sigma$  of 0.7 and 0.8, respectively. The value of the model in (c) $\chi^2$ = 1.3 is well within one $\sigma$ (0.8) and is thus a good fit. The model in (b) gives a $\chi^2$ value that is 7$\sigma$ away from the expected value. This fit is unacceptable based on the value of $\chi^2$.
